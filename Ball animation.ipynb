```python
from tkinter import *
import random as rd
maFenetre=Tk()
maFenetre.title('Affichage des balles')
canevas = Canvas(maFenetre, width = 480, height = 480, bg ='black')
canevas.pack(padx =5, pady =5,expand=True,fill=BOTH)
xc=240
yc=240
dx=3
r=20
dy=3
#prop=xc,yc,dx,dy,r
#objet=canevas.create_oval(xc-r, yc-r, xc+r, yc+r, fill='red')
couleurs=['red','blue','yellow','white','green','pink','purple','orange','brown']


booleen=True
bool_balles=True

class Balle_mobile:
    def __init__(self,parent,xc,yc,dx,dy,r,c):
        self.__xc=xc
        self.__yc=yc
        self.__r=r
        self.__c=c
        self.__dx=dx
        self.__dy=dy
        self.__parent=parent
        self.__objet=self.__parent.create_oval(xc-r, yc-r, xc+r, yc+r, fill=c)
        
    def move(self):
        self.__xc=self.__xc+self.__dx
        self.__yc=self.__yc+self.__dy
        
    def set_deplacement(self,dxx,dyy):
        self.__dx=dxx
        self.__dy=dyy
        
    def get_deplacement(self):
        return self.__dx,self.__dy
        
    def get_proprietes(self):
        return self.__xc,self.__yc,self.__dx,self.__dy,self.__r,self.__c
        
    def etre_au_bord(self):
        xc,yc,dx,dy,r,c=self.get_proprietes()
        w=self.__parent.winfo_width()
        h=self.__parent.winfo_height()
        if xc+r+dx>=w and yc+r+dy>=h:
            return True
        elif xc-r+dx<=0 and yc+r+dy>=h:
            return True
        elif xc+r+dx>=w and yc-r+dy<=0:
            return True
        elif xc-r+dx<=0 and yc-r+dy<=0:
            return True
        else:
            return False
    
    def verif_rebondir(self):
        xc,yc,dx,dy,r,c=self.get_proprietes()
        w=self.__parent.winfo_width()
        h=self.__parent.winfo_height()
        if xc+r+dx>=w and yc+r+dy<h:
            self.set_deplacement(-dx,dy)
            
        elif xc+r+dx<w and yc+r+dy>=h:
            self.set_deplacement(dx,-dy)
            
        elif xc-r+dx<=0 and yc-r+dy>0:
            self.set_deplacement(-dx,dy)
            
        elif xc-r+dx>0 and yc-r+dy<=0:
            self.set_deplacement(dx,-dy)
            
        elif self.etre_au_bord():
            self.set_deplacement(-dx,-dy)
        
                
            
    def deplacement(self):
        self.verif_rebondir()
        self.move()
        dxx,dyy=self.get_deplacement()
        self.__parent.move(self.__objet,dxx,dyy)
        
        

balles=[]

def stop_b():
    global booleen,boole
    booleen=False
    boole=True

def afficher():
    global booleen
    if booleen:
        if bool_balles and rd.random()<=(1/40):
            ajout_balles()
        for i in balles:
            i.deplacement()
        maFenetre.after(5,afficher)
    if booleen==False:
        booleen=True

boole=True        
def commencer():
    global boole
    if boole:
        boole=False
        afficher()
        
def ajout_balles():
    w=canevas.winfo_width()
    h=canevas.winfo_height()
    c2=rd.choice(couleurs)
    dx2=2*rd.random()+1
    dy2=2*rd.random()+1
    r=10*rd.random()+10
    x=r+(w-2*r)*rd.random()
    y=r+(h-2*r)*rd.random()
    new=Balle_mobile(canevas,x,y,dx2,dy2,r,c2)
    balles.append(new)
    
def stop_balles():
    global bool_balles
    bool_balles=False
    
Button(maFenetre,text='Afficher',command=commencer).pack(side=LEFT,padx=5,pady=5)
Button(maFenetre,text='Assez de balles',command=stop_balles).pack(side=LEFT,padx=5,pady=5)
Button(maFenetre,text='ArrÃªter',command=stop_b).pack(side=LEFT,padx=5,pady=5)    


maFenetre.mainloop()
```
